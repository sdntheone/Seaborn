## What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics, especially when working with structured datasets (e.g., Pandas DataFrames). Seaborn simplifies creating complex visualizations by automating tasks such as setting colors, adding statistical elements, and supporting thematic styles.

### Types of Charts in Seaborn

### 1. Relational Plots

Used to visualize relationships between variables.

* **Scatter Plot**<br>
Displays the relationship between two numerical variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset
data = pd.DataFrame({
    "age": [22, 25, 30, 35, 40],
    "income": [20000, 25000, 30000, 35000, 40000]
})

sns.relplot(x="age", y="income", data=data, kind="scatter")
plt.title("Scatter Plot: Age vs Income")
plt.show()
```

* **Line Plot**<br>
Shows trends or changes over time or an ordinal variable.

```python
sns.relplot(x="age", y="income", data=data, kind="line")
plt.title("Line Plot: Age vs Income")
plt.show()
```

### 2. Categorical Plots

Used to visualize relationships involving categorical data.

* **Bar Plot**<br>
    Shows aggregate values (e.g., mean) for each category.

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot: Average Bill per Day")
plt.show()
```

* **Box Plot**<br>
Displays the distribution of a dataset based on five summary statistics (min, Q1, median, Q3, max)

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot: Bill Distribution by Day")
plt.show()
```

* **Violin Plot**<br>
Combines a box plot with a kernel density estimate, showing both summary statistics and data distribution.

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot: Bill Distribution by Day")
plt.show()
```

* **Swarm Plot**<br>
Displays individual data points for a category, often used alongside box/violin plots.

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarm Plot: Bill Distribution by Day")
plt.show()
```

### 3. Distribution Plots

Used to visualize the distribution of a single variable.

* **Histogram**<br>
    Shows the frequency of data points in intervals.

```python
sns.histplot(tips["total_bill"], bins=10)
plt.title("Histogram: Bill Amount")
plt.show()
```

* **Kernel Density Estimate (KDE) Plot**<br>
Shows a smoothed curve representing the data's probability density.

```python
sns.kdeplot(tips["total_bill"], shade=True)
plt.title("KDE Plot: Bill Amount")
plt.show()
```

### 4. Matrix Plots

Used to visualize data in a grid-like structure.

* **Heatmap**<br>
Visualizes tabular data (e.g., correlation matrix) with color coding.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")
print(iris.head())  # Display the first few rows of the dataset
```

```python
corr = iris.drop("species", axis=1).corr()  # Exclude the categorical column
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

### 5. Pair Plot

Creates a grid of scatter plots and histograms for pairwise relationships in a dataset

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.suptitle("Pair Plot: Iris Dataset", y=1.02)
plt.show()
```

### 6. Regression Plots

Used to fit and visualize regression models.

* Simple Linear Regression

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot: Tip vs Total Bill")
plt.show()
```

## Some Important Parameter

### 1. hue Parameter

The `hue` parameter is used to group data by a categorical variable and apply different colors to the categories. This helps in visually distinguishing the different groups in the data.

**Explanation:**<br>
* `hue` takes a categorical column (e.g., `"species"`) and automatically assigns a different color to each category.
* This is especially useful when you want to visualize how a certain relationship behaves across different groups.

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
```

### 2. `cmap` Parameter

The `cmap` parameter is used to specify the color palette for a plot. It’s particularly useful in heatmaps, but can be used in various other plots like `scatterplot`, `lineplot`, etc.

**Explanation:**
* `cmap` stands for color map. It is a string representing the name of the color palette, or it can be a list of colors.
* Seaborn and Matplotlib offer a wide variety of built-in color maps, such as `coolwarm`, `viridis`, `Blues`, `RdBu`, etc.

```python
sns.heatmap(corr, annot=True, cmap="RdBu")
```

### 3.` bins` Parameter

The `bins` parameter is used to specify the number of bins (intervals) to divide the data into when plotting a histogram or ECDF (Empirical Cumulative Distribution Function).

**Explanation:**

* `bins` controls how the continuous data is grouped. A smaller number of bins means fewer intervals, and a larger number of `bins` means more precise intervals.
* Typically used in functions like `histplot()` or `displot()` for histograms.

```python
sns.histplot(iris["sepal_length"], bins=15, kde=True)
```

### 4. `kde` Parameter

The `Kernel Density Estimate (KDE)` is a way to estimate the probability density function (PDF) of a continuous variable. The `kde` parameter is used to specify whether to display the KDE curve alongside the histogram or other plots.

**Explanation:**

* `kde=True` adds the kernel density estimate curve to the plot.
* The KDE smooths out the data and shows the underlying distribution of the data. This is useful for visualizing the density of data and identifying patterns like bimodal distributions.

```python
sns.histplot(iris["sepal_length"], bins=10, kde=True)
```

Here, `kde=True` adds a KDE curve to the histogram. The KDE curve helps to visualize the smooth density of the sepal_length distribution.

***Without KDE:***

```python
sns.histplot(iris["sepal_length"], bins=10)
```

### 5. style

The style parameter adds different markers or line styles for categories in relational plots (e.g., `scatterplot`, `lineplot`).

**Explanation:**

This is useful when you need to distinguish groups not only by color but also by style.

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", style="species", data=iris)
```

* Different marker styles (circle, triangle, square, etc.) are applied for each species.

### 6. size

The size parameter allows you to control the size of points in plots based on another variable.

**Explanation:**

    Useful for representing a third numerical variable in a scatter plot.

```python
sns.scatterplot(x="sepal_length", y="petal_length", size="sepal_width", hue="species", data=iris, sizes=(20, 200))
```

Here, sepal_width determines the size of each point, with sizes=(20, 200) defining the range of point sizes.

### 7. linewidth

The `linewidth` parameter controls the width of lines in `line plots`, `scatter plots` (borders around markers), or bar plot outlines.

```python
sns.lineplot(x="sepal_length", y="petal_length", data=iris, linewidth=4)
```

### 5. alpha

The alpha parameter adjusts the transparency of the elements in the plot.

**Explanation:**

   * Values range from 0 (completely transparent) to 1 (fully opaque).
  * Useful for overlapping plots or reducing visual noise.

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, alpha=0.7)
```

### 6. legend

The legend parameter controls whether the legend is displayed and how it is styled.

**Explanation:**

You can turn off the legend using `legend=False` or modify its properties.

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris, legend="full")
```

### 7. aspect and height

These parameters are used in multi-plot grids (e.g., FacetGrid or catplot) to control the plot dimensions.

**Explanation:**

`aspect`: Ratio of width to height.<br>
`height`: Height of each subplot in inches

```python
sns.catplot(x="species", y="sepal_length", data=iris, kind="box", aspect=1, height=3)
```

### 8. markers

The markers parameter customizes the markers used in scatter or line plots.

**Explanation:**

Specify marker types for each category.<br>
Accepts a list of marker symbols (e.g., `"o"`, `"s"`, `"^"`).

```python
sns.lineplot(x="sepal_length", y="petal_length", hue="species", style="species", markers=["o", "s", "^"], data=iris)
```
